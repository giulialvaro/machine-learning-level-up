# KNN - Nearest Neighbors

# Librerías

1.Importamos las librerías necesarias

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
print ("listo!")
```

# Dataset

2. Creamos unos datos de prueba. data = sabemos la categoría y  punto_nuevo = no sabemos su género

```python
data = {'Masa': [50, 80, 90, 45, 60],
        'Altura': [1.48, 1.82, 1.85, 1.55, 1.60],
        'Genero': ['m', 'h', 'h', 'm', 'm']}
```

```python
punto_nuevo = {'Masa': [70],
               'Altura': [1.82]}
```

3. Convertimos a dataframe ...

```python
df = pd.DataFrame(data)
punto_nuevo = pd.DataFrame(punto_nuevo)
```

# Modelo

4. Graficamos los puntos!

```python
ax = plt.axes()

ax.scatter(df.loc[df['Genero'] == 'h', 'Masa'],
           df.loc[df['Genero'] == 'h', 'Altura'],
           c="red",
           label="Hombre")
ax.scatter(df.loc[df['Genero'] == 'm', 'Masa'],
           df.loc[df['Genero'] == 'm', 'Altura'],
           c="blue",
           label="Mujer")
ax.scatter(punto_nuevo['Masa'],
           punto_nuevo['Altura'],
           c="black")
plt.xlabel("Masa")
plt.ylabel("Altura")
ax.legend()
plt.show()
```

5. Entrenamos al algoritmo KNN con los datos para los que tenemos etiquetas

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
X = df[['Masa', 'Altura']]
y = df[['Genero']] #variable objetivo

knn.fit(X, y)
```

6. Vemos la predicción 

```python
prediccion = knn.predict(punto_nuevo)
print(prediccion)
```
