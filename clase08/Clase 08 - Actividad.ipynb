# Actividad de Clase üåª

```python
data = {
    'Tama√±o': [1500, 1600, 1700, 1800, 1900],
    'Habitaciones': [3, 3, 3, 4, 4],
    'Edad': [10, 15, 20, 25, 30],
    'Precio': [300000, 320000, 340000, 360000, 380000]
}
```

1. Librerias y dataframe

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.DataFrame(data)
df
```

2. Graficamos la relacion de cada var con el precio

```python
df.plot(x = "Tama√±o", y = "Precio", style = "o", color = "purple")
plt.title("Tama√±o vs Precio")
plt.xlabel("Tama√±o en m2")
plt.ylabel("Precio de la propiedad")
plt.show()
```

```python
df.plot(x = "Habitaciones", y = "Precio", style = "o", color = "purple")
plt.title("Habitaciones vs Precio")
plt.xlabel("Cantidad de habitaciones")
plt.ylabel("Precio de la propiedad")
plt.show()
```

```python
df.plot(x = "Edad", y = "Precio", style = "o", color = "purple")
plt.title("Antiguedad vs Precio")
plt.xlabel("Antiguedad de la propiedad")
plt.ylabel("Precio de la propiedad")
plt.show()
```

3. Divido en entrenamiento y prueba

```python
# variable indep y vars dep
x = df[["Tama√±o", "Habitaciones", "Edad"]]
y = df["Precio"]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.6, random_state=42)
```

4. Crear y entrenar el modelo

```python
# Creo el modelo
modelo = LinearRegression()

# Entreno el modelo
modelo.fit(X_train, y_train)
```

5. Hacer predicciones y evaluar el modelo

```python
# Predicciones con el conjunto de prueba
y_pred = modelo.predict(X_test)
print(y_pred)

# Evaluaci√≥n del modelo
MSE = mean_squared_error(y_test, y_pred)
R2 = r2_score(y_test, y_pred)

print(f"Error cuadr√°tico medio: {MSE}")
print(f"Coeficiente de determinaci√≥n: {R2}")
```

6. Visualizar resultados

```python
# Graficar la relaci√≥n entre cada caracter√≠stica y el precio
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(df['Tama√±o'], df['Precio'], color='purple')
plt.plot(df['Tama√±o'], modelo.predict(df[['Tama√±o', 'Habitaciones', 'Edad']]), color='pink')
plt.xlabel('Tama√±o')
plt.ylabel('Precio')
plt.title('Tama√±o vs Precio')

plt.subplot(1, 3, 2)
plt.scatter(df['Habitaciones'], df['Precio'], color='lightgreen')
plt.plot(df['Habitaciones'], modelo.predict(df[['Tama√±o', 'Habitaciones', 'Edad']]), color='pink')
plt.xlabel('Habitaciones')
plt.ylabel('Precio')
plt.title('Habitaciones vs Precio')

plt.subplot(1, 3, 3)
plt.scatter(df['Edad'], df['Precio'], color='yellow')
plt.plot(df['Edad'], modelo.predict(df[['Tama√±o', 'Habitaciones', 'Edad']]), color='pink')
plt.xlabel('Edad')
plt.ylabel('Precio')
plt.title('Edad vs Precio')

plt.tight_layout()
plt.show()
```
